```python
import os
import requests

try:
    from google.colab import userdata
except ImportError:
    userdata = None


def load_secret(name: str, default: str = "") -> str:
    if userdata is None:
        return os.getenv(name, default)
    try:
        return userdata.get(name)
    except Exception:
        return os.getenv(name, default)


api_base = load_secret("PREDICTION_API_BASE_URL", os.getenv("PREDICTION_API_BASE_URL", "")).rstrip("/")
if api_base:
    health = requests.get(f"{api_base}/api/health", timeout=20)
    print("Hosted API health:", health.status_code)
    print(health.json())
else:
    print("No hosted API base configured. Public showcase mode will use the local orchestrator instead.")
```

```python
from service.orchestrator import answer

queries = [
    "What is the latest housing pressure classification for Mayo?",
    "Compare Cork and Galway for affordable housing pressure.",
    "Explain homelessness pressure in Dublin in plain language.",
]

for q in queries:
    print("Q:", q)
    print("A:", answer(q))
    print("-" * 80)
```
